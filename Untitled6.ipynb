```

"""

Activation Functions are implemented:
1) Linear
2) ReLU
3) Leaky ReLU
4) Parametric ReLU
5) Sigmoid Function
6) TanH
7) Softmax
8) polynomial
"""

import numpy as np
import matplotlib.pyplot as plt
```

# **Linear Activation Function**

```
"""Linear Activation Function"""
x = np.arange(0,10,0.5)
plt.axhline(0,color='red') 
plt.axvline(0,color='red') 
plt.plot(x)
plt.grid()
```

# **Rectified linear activation function (ReLu)**

```
""" ReLu"""

def reLu(x):
  return np.array([xi if xi >= 0 else 0 for xi in x])
```

```
x=np.arange(-2,2,0.01)
plt.axhline(0,color='red') 
plt.axvline(0,color='red') 
plt.plot(x, reLu(x))
plt.grid()
```

# **Exponential Linear Unit (ELU)**

```
"""Exponential Linear Unit (ELU)"""
def eLU(x, alpha=1):
  return np.array([xi if xi>0 else alpha*(np.exp(xi)-1) for xi in x ])
```

```
x=np.arange(-10,10,0.01)
plt.axhline(0,color='red') 
plt.axvline(0,color='red') 
plt.plot(x, eLU(x))
plt.grid()
```

# **Sigmoid / Logistic Activation Function**

```
"""Sigmoid / Logistic Activation Function"""
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```
x=np.arange(-10,10,0.005)
plt.axhline(0,color='red') 
plt.axvline(0,color='red') 
plt.plot(x, sigmoid(x))
plt.grid()
```

# **logsigmoid**

```
"""logsigmoid"""
def logsigmoid(x, lam=1):
  return np.log(1/(1+np.exp(-lam*x)))
```

```
x=np.arange(-5,5,0.01)
plt.axhline(0,color='red') 
plt.axvline(0,color='red') 
plt.plot(x, logsigmoid(x))
plt.grid()
```

# **TanH / Hyperbolic Tangent Activation Function**

```
"""TanH / Hyperbolic Tangent Activation Function"""

def tanh(x):
    a = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    return a

y = tanh(x)
plt.plot(x,y)
plt.grid()
```

# **Softmax Activation Function**

```

"""Softmax Activation Function"""
def softmax_function(x):
    z = np.exp(x)
    m = z/z.sum()
    return m

y = softmax_function(x)
plt.plot(x,y)
plt.grid()
```
